## SMOTE (Synthetic Minority Oversampling Technique)
SMOTE is a technique used in machine learning to address embalanced datasets where the minority class is significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(
    n_samples=1000,         # Total number of samples (rows) in the dataset
    n_features=2,           # Total number of features (columns) for each sample
    n_redundant=0,          # Number of redundant features (none in this case)
    n_clusters_per_class=1, # Number of clusters per class
    weights=[0.90],         # Class weights; 90% of samples will belong to one class
    random_state=12         # Seed for reproducibility
)

# Xis independent fearure y is dependent fearure
```

```python
import pandas as pd
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df.target.value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])

#c=final_df['target'] means that the color of each point
```
